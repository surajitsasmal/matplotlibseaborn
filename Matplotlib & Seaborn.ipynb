# Lineplot

```python
#It is a simple plot that displays the relationship between two variables
#To see  the trend in the dataset we use lineplot
```

```python
month=['jan','Feb','March','April','may','june']
price=[100,234,256,456,456,678]
plt.plot(month,price,color="b",marker="*")
plt.xlabel("month")
plt.ylabel('price')
plt.title("month vs average stock price")
plt.grid()
plt.show()
```

## Multiple Lineplot

```python
plt.figure(figsize=(9,5))
month=['jan','Feb','March','April','may','june']
furniture_sales=[1000,2000,2345,1200,3456,4567]
book_sales=[2345,1234,4567,3456,1000,1234]
tech_sales=[2345,3678,1890,19000,7890,5678]
plt.plot(month,furniture_sales,color='b',marker='*',label='furniture_sales')
plt.plot(month,book_sales,color='g',marker='*',label='book_sales')
plt.plot(month,tech_sales,color='r',marker='*',label='tech_sales')
plt.xlabel("month")
plt.ylabel("item sales")
plt.title("various item sales in each item")
plt.legend()
plt.show()
```

# Barplot

```python
#It is used to display the categorical data with bars of lengths proportional to the values that they represent 
```

```python
import seaborn as sns
df_tips = sns.load_dataset('tips')
df_tips.head()
```

```python
plt.figure(figsize=(15,8))

df_tips.smoker.value_counts().plot(kind='bar', color = 'b')

plt.text(x = -0.06, y = df_tips.smoker.value_counts()[1]+59, s = df_tips.smoker.value_counts()[1])
plt.text(x = 0.98, y = df_tips.smoker.value_counts()[0]-40, s = df_tips.smoker.value_counts()[0])

plt.title('Bar Plot')

plt.xlabel('Smoker category')

plt.ylabel('Count of smoker')

plt.show()
```

# Group_bar_chart

```python
import numpy as np
p_score=[10,12,13,15]
s_score=[13,15,16,18]
index=np.arange(len(p_score))
plt.bar(x=index,height=p_score,width=0.35,label='p_score')
plt.bar(x=index+0.35,height=s_score,width=0.35,label='s_score')
plt.title("Exam score")
plt.xlabel("subject")
plt.ylabel("score")
plt.xticks(ticks=index+0.35/2,labels=['A','B','C','D'])
plt.legend()
plt.show()
```

# Horizontal_bar_chart

```python
y=['jan', 'Feb', 'march', 'april', 'may']
 
# getting values against each value of y
x=[50,24,135,67,122]
plt.barh(y, x)
 
# setting label of y-axis
plt.ylabel("Month")
 
# setting label of x-axis
plt.xlabel("price")
plt.title("Horizontal bar graph")
plt.show()
```

# stacked_bar_chart

```python
p_score=[10,12,13,15]
s_score=[13,15,16,18]
index=np.arange(len(p_score))
plt.bar(x=index,height=p_score,width=0.80,label='p_score')
plt.bar(x=index,height=s_score,width=0.80,bottom=p_score,label='s_score')
plt.title("Exam score")
plt.xlabel("subject")
plt.ylabel("score")
plt.xticks(ticks=index+0.35/2,labels=['A','B','C','D'])
plt.legend()
plt.show()
```

# Histogram

It is used to represent the distribution of the numeric variable.

It is an estimate of the probability distribution of a continuous data.

One axis represents the variable in the form of bars and another represents the frequency each bar


```python
plt.hist(df_tips['tip'], bins= 5)
 
plt.title('Distribution of tip amount')

plt.xlabel('tip')
plt.ylabel('Frequency')
plt.show()
```

# Boxplot

 used to visualize the distribution of the numeric variable.

Represents the five number summary of the variable which includes the minimum,

first quartile (Q1), second quartile (median), third quartile (Q3) and maximum of the variable

Used to detect the outliers  in the data


```python
plt.boxplot(df_tips['total_bill'])

plt.text(x = 1.1, y = df_tips['total_bill'].min(), s ='min')             
plt.text(x = 1.1, y = df_tips.total_bill.quantile(0.25), s ='Q1')         
plt.text(x = 1.1, y = df_tips['total_bill'].median(), s ='median (Q2)')   
plt.text(x = 1.1, y = df_tips.total_bill.quantile(0.75), s ='Q3')         
plt.text(x = 1.1, y = df_tips['total_bill'].max(), s ='max')             

plt.title('Boxplot of Total Bill Amount')

plt.ylabel('Total bill')

plt.show()
```

```python
plt.figure(figsize =(10, 7))
data = np.random.normal(100, 20, 200)

plt.boxplot(data)

plt.show()
```

# Distribution_plot

A Seaborn displot or distribution plot depicts the univariate or bivariate distribution of the data in a

dataset. It combines a histogram with optional components,

such as a Kernel Density Estimation (KDE) line or rug plot.

```python
sns.set_style('whitegrid')
sns.distplot(df_tips['total_bill'], kde = False, color ='red', bins = 30)
```
